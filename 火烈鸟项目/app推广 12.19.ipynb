```python
#读取数据

#data_11 = pd.read_csv('11月订单.csv',engine = 'python')[['UIN','订单金额','归属公会账号','首次购买时间']]
data_12 = pd.read_csv('12月订单.csv',engine = 'python')
data_table = pd.read_csv('用户首次app购买列表1576831902.csv',engine = 'python')
data_12 = pd.merge(data_12,data_table,left_on = 'UIN',right_on = '用户uin',how = 'left')[['UIN','订单金额','归属公会账号','首次购买日期']]
data_12 = data_12[data_12['首次购买日期']>= '2019-12-01 00:00:00']
```

```python
data = data_12.groupby(['UIN','归属公会账号']).sum().reset_index()
data['区间'] = pd.cut(data['订单金额'],[0,6],right = True)
data2 = data.groupby(['区间','归属公会账号']).agg([np.sum,'count']).dropna()['订单金额']

data3 = data_12.groupby('归属公会账号').agg([np.sum,'count'])['订单金额'].reset_index()
data4 = pd.merge(data2.reset_index(),data3,on = '归属公会账号',how = 'left')
data4['金额占比'] = data4['sum_x'] / data4['sum_y']
data4['数量占比'] = data4['count_x'] / data4['count_y']
result = data4.sort_values('数量占比',ascending = False)

table = pd.read_csv('推广活动参与列表1576835617.csv',engine = 'python')
table = table[table['活动名称']=='新增用户在APP自助充值奖励A']
table = table[table['当前人数']>=20]
result
```
