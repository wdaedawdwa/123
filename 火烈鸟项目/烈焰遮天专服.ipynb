```python
#游戏id与总流水
data_201907 = pd.read_excel('7.31数据源.xlsx')[['游戏ID','本月新增流水','本月总流水']].groupby('游戏ID').sum()
data_201908 = pd.read_excel('8.31数据源.xlsx')[['游戏ID','本月新增流水','本月总流水']].groupby('游戏ID').sum()
data_201909 = pd.read_excel('9.30数据源.xlsx')[['游戏ID','本月新增流水','本月总流水']].groupby('游戏ID').sum()
data_201910 = pd.read_excel('10.31数据源.xlsx')[['游戏ID','本月新增流水','本月总流水']].groupby('游戏ID').sum()
data_201911 = pd.read_excel('11.30数据源.xlsx')[['游戏ID','本月新增流水','本月总流水']].groupby('游戏ID').sum()
data_201912 = pd.read_excel('12.31数据源.xlsx')[['游戏ID','本月新增流水','本月总流水']].groupby('游戏ID').sum()
data_202001 = pd.read_excel('1.31数据源.xlsx')[['游戏ID','本月新增流水','本月总流水']].groupby('游戏ID').sum()
data_202002 = pd.read_excel('2.29数据源.xlsx')[['游戏ID','本月新增流水','本月总流水']].groupby('游戏ID').sum()
```

```python
data_sum = pd.concat([data_201907,data_201908,data_201909,data_201910,data_201911,data_201912,data_202001,data_202002]).reset_index().groupby('游戏ID').sum()
```

```python
#游戏类型
game_type = pd.read_csv('大事件.csv',engine = 'python')
game_type = game_type[(game_type['主玩法'] == '传奇类MMO')|game_type['游戏名称'].str.contains('传奇')][['App ID','游戏名称']]
game_type.dropna(inplace = True)
game_type.head()
```

```python
result = data_sum.sort_values('本月总流水',ascending = False)
result = pd.merge(result,game_type,left_index = True,right_on = 'App ID',how = 'left')
result.dropna().head()
```
