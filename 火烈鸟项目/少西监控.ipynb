```python
'''
(1)66+果盘双端数据
'''

lst = os.listdir('C:/Users/Administrator/Desktop/工具包/日常监控/')
m = []

for i in lst:
    data = pd.read_csv(i,engine = 'python').sort_values('日期').groupby('日期').sum()
    data_select = data[[data.columns[2],data.columns[5],data.columns[6],data.columns[7]]]
    m.append(data_select)
    
a = pd.concat([m[0],m[1],m[2],m[3]],axis = 1)
a.columns = ['新增用户a','新增付费用户a','新增付费用户收入a','收入（折扣前）a',
             '新增用户b','新增付费用户b','新增付费用户收入b','收入（折扣前）b',
             '新增用户c','新增付费用户c','新增付费用户收入c','收入（折扣前）c',
             '新增用户d','新增付费用户d','新增付费用户收入d','收入（折扣前）d',]
a['新增汇总'] = a['新增用户a']+a['新增用户b']+a['新增用户c']+a['新增用户d']
a['新增付费汇总'] = a['新增付费用户a']+a['新增付费用户b']+a['新增付费用户c']+a['新增付费用户d']
a['新增付费用户收入汇总'] = a['新增付费用户收入a']+a['新增付费用户收入b']+a['新增付费用户收入c']+a['新增付费用户收入d']
a['收入（折扣前）汇总'] = a['收入（折扣前）a']+a['收入（折扣前）b']+a['收入（折扣前）c']+a['收入（折扣前）d']

a[['新增汇总','新增付费汇总','新增付费用户收入汇总','收入（折扣前）汇总']]
```

```python
'''

(2)果盘公会数据明细

'''

match_table = pd.read_excel('C:/Users/Administrator/Desktop/渠道对接人1210.xlsx')

print(lst)
m = [lst[2],lst[3]]

a = []
for i in m:
    #print(i)
    datai = pd.read_csv(i,engine = 'python')
    data_matchi = pd.merge(datai,match_table,on = '渠道ID',how = 'left')[['日期','会长账号','新增用户','新增付费用户','新增付费用户收入','收入']]
    a.append(data_matchi.groupby(['会长账号','日期']).sum())
    
```

```python
n = 0

df_sum = a[n].reset_index().groupby('会长账号').sum().reset_index()
df_sum['日期'] = '汇总'
df_sum.sort_values('收入',ascending = False,inplace = True)

df = a[n].reset_index()


df_c = pd.concat([df_sum,df]).groupby(['会长账号','日期']).sum()
#df_c
```
