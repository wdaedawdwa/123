```python
#分割时间
data = pd.read_csv('霸王雄心11.5-11.18.csv',engine = 'python')
data['交易时间'] = pd.to_datetime(data['交易时间'])
data.index = data['交易时间']
data['date'] = data.index.day
data = data.sort_values('date',ascending = True)

#数据分组
m = []
for day in data['date'].unique().tolist():
    data_day = data[data['date'] == day]
    data_g = data_day.groupby('用户名(66UIN)').sum()
    data_g['区间'] = pd.cut(data_g['原价'],bins = [0,100,500,1000,3000,999999],right = False)
    data_gg = data_g.groupby('区间').agg([len,np.sum])
    data_gg.fillna(0,inplace = True)
    data_gg.index = [0,100,500,1000,'≥3000']
    m.append(data_gg[['原价']])

m[13]
```

```python
#渠道
data2 = pd.read_csv('xxStatistics_66SDK（66平台）_所有渠道_122004_SDK游戏商业分析_20191105至20191118.csv',engine = 'python')

data2_Mdate = data2[['渠道ID','日期']].groupby('渠道ID').min()
data_id = data2[['渠道ID','渠道名','收入（折扣前）','新增用户']].groupby(['渠道ID','渠道名']).sum().reset_index()

data_m = pd.merge(data_id,data2_Mdate,left_on = '渠道ID',right_index = True,how = 'left')

data_m['起量日期'] = data_m['日期'] - 20191100
data_m['起量天数'] = 18 + 1 - data_m['起量日期'] 
data_m['日均新增'] = data_m['新增用户'] / data_m['起量天数']
data_m['日均流水'] = data_m['收入（折扣前）'] / data_m['起量天数']
data_m = data_m.sort_values('日均流水',ascending = False)

data_m.head(10)
```
