```python
data = pd.read_csv('2月用研数据底表.csv',engine = 'python').sort_values('日期',ascending = True)
data.head()
```

```python
#基本数据

data['月份'] = data['日期'].astype(str).str[4:6] + '月'

data_12 = data[data['月份'] == '12月']
data_01 = data[data['月份'] == '01月']

print(data_12['收入'].describe())
print('-----')
print(data_01['收入'].describe())
```

```python
from numpy import arange

#plt.style.use('fivethirtyeight')

data_12['收入'].plot.kde(legend = True,label = 201912)
data_01['收入'].plot.kde(legend = True,label = 202001)

#plt.xlabel('收入',fontproperties = 'SimHei',fontsize = 11)
#plt.ylabel('分布概率',fontproperties = 'SimHei',fontsize = 11)
plt.xlim(0,550000) # because ratings start at 0 and end at 5
plt.xticks(arange(0,550000,50000))
plt.show()
```
