```python
data = pd.read_csv('1.csv',engine = 'python')
data['定基增长量'] = data['新增流水'] - data['8月新增流水均值']
data['定基增长率'] = data['定基增长量'] / data['新增流水']
data['逐期增长量'] = data['新增流水'] - data['新增流水'].shift(1)
data['逐期增长率'] = data['逐期增长量'] / data['新增流水'].shift(1)
data.fillna(0,inplace = True)

data[['定基增长量','逐期增长量']].plot(figsize = (10,4),style = '--.',alpha = 0.8)
plt.axhline(0,hold = None,color = 'r',linestyle = '--')



data
```
