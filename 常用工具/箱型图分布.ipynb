```python
data = pd.read_csv('少三安卓.csv',engine = 'python')
data = data[['日期','收入','付费用户']]
data['月份'] = '暂未确定'
data['月份'][data['日期'] < 20191201 ] = '11_month'
data['月份'][data['日期'] >= 20191201 ] = '12_month'
data = data.sort_values('日期',ascending = True)
data.columns = ['date','income','Paying users','month']

sns.boxplot(x = 'month',y = 'income',data = data,
           palette = 'Reds')

q1_first = data[data['month'] == '11_month']['income'].quantile(q = 0.25)
q2_first = data[data['month'] == '11_month']['income'].quantile(q = 0.5)
q3_first = data[data['month'] == '11_month']['income'].quantile(q = 0.75)

q1_second = data[data['month'] == '12_month']['income'].quantile(q = 0.25)
q2_second = data[data['month'] == '12_month']['income'].quantile(q = 0.5)
q3_second = data[data['month'] == '12_month']['income'].quantile(q = 0.75)

print(q1_first,q2_first,q3_first)
print(q1_second,q2_second,q3_second)
```
