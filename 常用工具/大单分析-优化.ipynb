```python
#复用参数

road = 'C:/Users/Administrator/Desktop/项目/大单研究/少三安卓/'
data_name = '公会发放果币卡记录1569226868.csv'
first_charge_point = 35
```

```python
import numpy as np
import pandas as pd
import datetime
import os
os.chdir(road)
pd.set_option('display.float_format',lambda x : '%.2f' % x)
```

```python
#数据读取与合并

data = pd.read_csv(data_name,engine = 'python')[['用户uin','代金券金额','消耗果粒',
                                                                 '发放时间','公会ID']]

data['发放时间'] = pd.to_datetime(data['发放时间'])
data['折扣'] = data['消耗果粒'] / data['代金券金额']*100
data['发放月份'] = data['发放时间'].map(lambda x: x.month)        #提取月份

data['折扣'] = data['折扣'].astype('float64')

#data = data.loc[data['公会ID'] == 77362]
```

```python
def Nusers_pay_thismonth(df):
    m = []
    for i in range(data['发放月份'].max()+1)[1:]:
        df_sum = df.loc[df['发放月份'] == i]
        df_sump = df_sum.groupby('用户uin').sum()
        df_first = df_sum.loc[df['折扣'] == first_charge_point]
        df_last = pd.merge(df_first,df_sump,left_on = '用户uin',right_index = True)[['用户uin','代金券金额_x',
                                                                                    '代金券金额_y','发放月份_x',]]
        df_last.rename(columns = {'代金券金额_x':str(i)+'月'+'首充','代金券金额_y':str(i)+'月'+'总值'},inplace = True)
        df_last.sort_values(df_last.columns[1],ascending = False,inplace = True)
        
        #Feb
        df_more1 = df.loc[df['发放月份']==i+1]
        df_more1p = df_more1.groupby('用户uin').sum()
        df_last = pd.merge(df_last,df_more1p,left_on = '用户uin',right_index = True,how='left')[['用户uin',df_last.columns[1],
                                                                                    df_last.columns[2],'代金券金额']]
        df_last.rename(columns = {'代金券金额':str(i+1)+'月'+'总值'},inplace = True)
        
        #Mar
        df_more2 = df.loc[df['发放月份']==i+2]
        df_more2p = df_more2.groupby('用户uin').sum()
        df_last = pd.merge(df_last,df_more2p,left_on = '用户uin',right_index = True,how='left')[['用户uin',df_last.columns[1],
                        df_last.columns[2],df_last.columns[3],'代金券金额']]
        df_last.rename(columns = {'代金券金额':str(i+2)+'月'+'总值'},inplace = True)
        
        #April
        df_more3 = df.loc[df['发放月份']==i+3]
        df_more3p = df_more3.groupby('用户uin').sum()
        df_last = pd.merge(df_last,df_more3p,left_on = '用户uin',right_index = True,how='left')[['用户uin',df_last.columns[1],
                        df_last.columns[2],df_last.columns[3],df_last.columns[4],'代金券金额']]
        df_last.rename(columns = {'代金券金额':str(i+3)+'月'+'总值'},inplace = True)
        
        #May
        df_more4 = df.loc[df['发放月份']==i+4]
        df_more4p = df_more4.groupby('用户uin').sum()
        df_last = pd.merge(df_last,df_more4p,left_on = '用户uin',right_index = True,how='left')[['用户uin',df_last.columns[1],
                        df_last.columns[2],df_last.columns[3],df_last.columns[4],df_last.columns[5],'代金券金额']]
        df_last.rename(columns = {'代金券金额':str(i+4)+'月'+'总值'},inplace = True)
        
        #Jun
        df_more5 = df.loc[df['发放月份']==i+5]
        df_more5p = df_more5.groupby('用户uin').sum()
        df_last = pd.merge(df_last,df_more5p,left_on = '用户uin',right_index = True,how='left')[['用户uin',df_last.columns[1],
                        df_last.columns[2],df_last.columns[3],df_last.columns[4],df_last.columns[5],
                        df_last.columns[6],'代金券金额']]
        df_last.rename(columns = {'代金券金额':str(i+5)+'月'+'总值'},inplace = True)
        
        #July
        df_more6 = df.loc[df['发放月份']==i+6]
        df_more6p = df_more6.groupby('用户uin').sum()
        df_last = pd.merge(df_last,df_more6p,left_on = '用户uin',right_index = True,how='left')[['用户uin',df_last.columns[1],
                        df_last.columns[2],df_last.columns[3],df_last.columns[4],df_last.columns[5],
                        df_last.columns[6],df_last.columns[7],'代金券金额']]
        df_last.rename(columns = {'代金券金额':str(i+6)+'月'+'总值'},inplace = True)
        
        #Aug
        df_more7 = df.loc[df['发放月份']==i+7]
        df_more7p = df_more7.groupby('用户uin').sum()
        df_last = pd.merge(df_last,df_more7p,left_on = '用户uin',right_index = True,how='left')[['用户uin',df_last.columns[1],
                        df_last.columns[2],df_last.columns[3],df_last.columns[4],df_last.columns[5],
                        df_last.columns[6],df_last.columns[7],df_last.columns[8],'代金券金额']]
        df_last.rename(columns = {'代金券金额':str(i+7)+'月'+'总值'},inplace = True)
        
        #Set
        df_more8 = df.loc[df['发放月份']==i+8]
        df_more8p = df_more8.groupby('用户uin').sum()
        df_last = pd.merge(df_last,df_more8p,left_on = '用户uin',right_index = True,how='left')[['用户uin',df_last.columns[1],
                        df_last.columns[2],df_last.columns[3],df_last.columns[4],df_last.columns[5],
                        df_last.columns[6],df_last.columns[7],df_last.columns[8],df_last.columns[9],'代金券金额']]
        df_last.rename(columns = {'代金券金额':str(i+8)+'月'+'总值'},inplace = True)
        
        
        
        df_last.fillna(0,inplace = True)
        m.append(df_last)
    return(m)

#Nusers_pay_thismonth(data)[8].sort_values('9月首充',ascending = False)
```

```python
#最终结果

def f1(lst):
    m = []
    for df in lst:
        cut = pd.cut(df[df.columns[1]],[0,5000,8000,10010,9999999999],right = False)
        df['区间'] = cut.values
        df = df.groupby('区间').sum()
        df['人数'] = cut.value_counts()
        df.fillna(0,inplace = True)
        m.append(df)
    return(m)

f1(Nusers_pay_thismonth(data))[8]
```
