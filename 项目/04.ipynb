```python
#数据读取
df = pd.read_csv('爱奇艺视频数据.csv',engine= 'python')

#数据清洗
df_c = f1(df)
df_c = f2(df_c,'数据获取日期')


#问题一：分析出不同导演电影的好评率，并筛选出TOP20
df_q1 = df_c.groupby('导演').sum()[['好评数','评分人数']]
df_q1['好评率'] = df_q1['好评数']/df_q1['评分人数']
df_q1_top20 = df_q1.sort_values('好评率',ascending = False).head(20)

#问题二：统计分析2001-2016年每年评影人数总量
```

```python
def f1(df):
    for i in df.columns:
        if df[i].dtype == 'object':
            df[i].fillna('缺失数据',inplace = True)
        else:
            df[i].fillna(0,inplace = True)
    return(df)

def f2(df,*cols):
    for col in cols:
        df[col] = df[col].str.replace('年','.')
        df[col] = df[col].str.replace('月','.')
        df[col] = df[col].str.replace('日','.')
        df[col] = pd.to_datetime(df[col])
    return(df)
```

```python
df_q2 = df_c[['导演','上映年份','整理后剧名']].drop_duplicates()  
df_q2 = df_q2[df_q2['上映年份'] != 0]

df_q2_v = df_c.groupby('整理后剧名').sum()[['评分人数','好评数']]

df_q2_m = pd.merge(df_q2,df_q2_v,left_on = '整理后剧名',
                   right_index = True,how = 'inner')

df_q2_s = df_q2_m.groupby('上映年份').sum()[['评分人数','好评数']]

print(df_q2_s.loc[2000:])

```
